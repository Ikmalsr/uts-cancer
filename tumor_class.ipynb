```python
df=df[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
features = ['radius_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','fractal_dimension_mean', 'texture_mean']
X = df[features]
Y = df['diagnosis']
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
sd_data = scaler.transform(X)
```

```python
X = sd_data
Y = df['diagnosis']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
x_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(x_train_prediction, Y_train)
```

```python
print('Akurasi data training = ', training_data_accuracy)
```

```python
x_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(x_test_prediction, Y_test)
```

```python
print('Akurasi data test = ', test_data_accuracy)
```
